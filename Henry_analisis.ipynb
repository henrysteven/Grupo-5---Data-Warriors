```python
df_heap = df.loc[:,['NIVEL DETALLADO','AREA DE ESTUDIO','MONTO CONTRATADO','ID UNICO']]
df_heap
```

```python
df_pivot = df_heap.groupby(['AREA DE ESTUDIO','NIVEL DETALLADO'])['MONTO CONTRATADO'].sum().reset_index()
df_pivot
```

```python
columns = 'AREA DE ESTUDIO'
row = 'NIVEL DETALLADO'
heatmap1_data = pd.pivot_table(df_pivot, values='MONTO CONTRATADO', 
                     index=[columns], 
                     columns=row)
```

```python
colors=["#22a7f0", "#63bff0", "#a7d5ed", "#e2e2e2", "#e1a692", "#de6e56", "#e14b31", "#c23728"]
sns.set()
sns.set(rc={'figure.figsize':(11.7,8.27)},font_scale = 1.25)
p=sns.heatmap(heatmap1_data, cmap="YlGnBu")
_ = p.set_ylabel(columns)
_ = p.set_xlabel(row)
_ = p.set_title("HeatMap de Área de estudio vs Nivel")
```
